```python
import cv2
import numpy as np

image = "/Users/frbuccoliero/Desktop/Thesis/Tests/template_matching/dataset/images/positive/positive_00000.jpg"

# Load the image
img = cv2.imread(image)

# Extract features with SIFT
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(img, None)

```

```python
import matplotlib.pyplot as plt

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 1. Original image
original = img_rgb.copy()

# 2. Original image with centers of keypoints (dots of equal size)
centers_img = img_rgb.copy()
for kp in keypoints:
    center = (int(kp.pt[0]), int(kp.pt[1]))
    cv2.circle(centers_img, center, 10, (255, 0, 0), -1)  # blue dot

# 3. Original image with keypoints drawn with size and orientation
keypoints_img = cv2.drawKeypoints(
	img_rgb, keypoints, None, color=(255, 0, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

# Plotting all three
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
axs[0].imshow(original)
axs[0].set_title("Original Image")
axs[0].axis('off')

axs[1].imshow(centers_img)
axs[1].set_title("Features")
axs[1].axis('off')

axs[2].imshow(keypoints_img)
axs[2].set_title("Features (Size & Orientation)")
axs[2].axis('off')

# Show and save
plt.tight_layout()
plt.savefig("sift_visualization.png")
plt.show()
```

```python
from bs4 import BeautifulSoup as bs
import requests
from pyWikiCommons import pyWikiCommons
```

# Download traffic signs

```python
url = 'https://commons.wikimedia.org/wiki/Road_signs_in_Italy'
```

```python
# Find all the <a> with class mw-file-description
response = requests.get(url)
soup = bs(response.text, 'html.parser')
links = soup.find_all('a', class_='mw-file-description')

# Get the href, e.g. : /wiki/File:Italian_traffic_signs_-_segni_orizzontali_in_rifacimento_(modello_II_6-a).svg
# and keep the part starting from "File:" to use in the download method

toDownload = []

for link in links:
    href = link.get('href')
    if href and 'File:' in href:
            filename = href.split('File:')[1]
            if filename != 'No_image_available.svg':
              toDownload.append(f"File:{filename}")
              print(f'Added {filename}')
```

```python
for file in toDownload:
    pyWikiCommons.download_commons_image(file)
```

# Download ISO signals
Please note that this only downloads the ISO 7010:2019 signals, while other from older versions (2003, 2011) are likely to be found in real life. To find those please refer to the ISO standard

```python
symbols = {
	"E":70,
	"F":19,
	"M":62,
	"P":75,
	"W":80
}

for s in symbols:
	for n in range(1,symbols[s]+1):
		full_name = f"File:ISO_7010_{s}{'0' if n < 10 else ''}{'0' if n < 100 else ''}{n}.svg"
		try:
			pyWikiCommons.download_commons_image(full_name)
		except:
			print(f"Error downloading {full_name}")
```
